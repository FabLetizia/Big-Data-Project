```python
import matplotlib.pyplot as plt
```

```python
def generate_graphical(tempi_locale, tempi_aws):
    dimensioni_input = [50, 100, 150, 200]
    etichette_dimensioni = ['50% (1.4 GB)', '100% (2.81 GB)', '150% (4.21 GB)', '200% (5.62 GB)']

    # Graphical
    plt.figure(figsize=(10, 6))
    plt.plot(dimensioni_input, tempi_locale, marker='o', label='Locale')
    plt.plot(dimensioni_input, tempi_aws, marker='o', label='AWS')

    # Labels and title
    plt.title('Tempi di esecuzione: Locale e AWS')
    plt.xlabel('Dimensione Input (%)')
    plt.ylabel('Tempo (sec)')
    plt.legend()
    plt.xticks(dimensioni_input, etichette_dimensioni)
    plt.ylim(bottom=0)

    plt.grid(True)
    return plt
```

# JOB 1

### HADOOP

```python
''' 
tempo esecuzione:
LOCALE:
50%: 78 sec
100%: 172 sec
150%: 243 sec
200%: 298 sec

AWS:
 50%: 45 sec
 100%: 85 sec
 150%: 86 sec
 200%: 119 sec 
 '''

tempi_locale = [78, 172, 243, 298]
tempi_aws = [45, 85, 86, 119]

plt = generate_graphical(tempi_locale, tempi_aws)
plt.show()
```
